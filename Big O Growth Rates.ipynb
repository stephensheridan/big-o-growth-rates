```python
import math
import matplotlib.pyplot as plt
import numpy
%pylab inline
```

```python
# Let's generate some increasing numbers to work with Interval 1 - 20 in steps of 0.25
values = numpy.arange(1, 20.25, 0.25)

# Print out the values in values
for x in values:
    print x,
```

```python
# x alone will give us O(n)
plt.plot(values,[x for x in values], label="O(n)")
# x*x will give us O(x^2)
plt.plot(values,[x*x for x in values], label="O(n^2)")
# x*x*x will give us O(x^3)
plt.plot(values,[x*x*x for x in values], label="O(n^3)")
# x * math.log(x,2) will give us O(n * log2(n))
plt.plot(values,[x * math.log(x,2) for x in values], label="O(n * log2(n))")
# log2(x) will give us O(log2(n))
plt.plot(values,[math.log(x,2) for x in values], label="log2(n)")
plt.ylim((1,100))
plt.xlim((1,20))
pylab.legend(loc='upper right')
```
